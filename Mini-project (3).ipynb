```python
# Extract the hour from the start_time
original_df['start_hour'] = original_df['Start Time'].dt.hour
original_df['Duration'] = (original_df['End Time'] - original_df['Start Time']).dt.total_seconds() / 3600
```

```python
def classify_time(hour):
    if 0 <= hour < 6:
        return 'Dawn'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'
```

```python
original_df['time_slot'] = original_df['start_hour'].apply(classify_time)
```

```python
original_df
```

```python
pivot_df = original_df.pivot_table(index='Name', columns='time_slot', values='Duration', aggfunc='sum', fill_value=0)

# Plot the grouped bar chart
ax = pivot_df.plot(kind='bar', figsize=(12, 7), width=0.8)

# Customize the chart
plt.title('Study Duration by Time Slot for Each Student', fontsize=16)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Duration (hours)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.legend(title='Time Slot', fontsize=12)
plt.tight_layout()

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge', fontsize=10)

# Show the plot
plt.show()
```
